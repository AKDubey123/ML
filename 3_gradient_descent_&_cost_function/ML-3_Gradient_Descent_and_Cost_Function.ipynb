Let's implement gradient descent funtion for the following<br>
__Linear equation__
y = 2x + 3<br>
x = [1,2,3,4,5]<br>
<font color=red>__y = [5,7,9,11,13]__</font> 



```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
y
```

```python
%matplotlib inline
# this funtion is to reduce the cost
def gradient_descent(x,y):
    
    #let start with randaom values of 'm' & 'b' nothing but slope/coefficient & intercept values respectively.
    m_curr = b_curr = 0
    iteration = 1000
    n = len(x)
    learning_rate = 0.05 # need to adjust to get global minimaa
    
    plt.scatter(x,y,color='red',marker='+',linewidths=10)
    plt.xlabel('x',fontsize=20)
    plt.ylabel('y',fontsize=20)
    plt.grid()
    plt.legend(title='linear equation\ny = m * x + b\ny = 2 * x + 3',loc='upper left')
    plt.title('Liner Equation - Gradient Descent',fontsize=16)
    
    # compare whether two values are close to each other, or not
    import math
    mse_old = 0.0
    
    for i in range(iteration):
        #linear equation => y = m * x + b
        y_predicted = m_curr * x + b_curr
        plt.plot(x,y_predicted,color='green')
        
        # mse calculation
        mse = (1/n) * sum([val ** 2 for val in (y - y_predicted)])
        if(math.isclose(mse_old,mse,abs_tol=0.01)):
            break
        mse_old = mse
        
        #partial derivative calculation
        md = -(2/n) * sum(x * (y -y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        #step learning rate
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {}, b {}, mse {}, iteration {}".format(m_curr,b_curr,mse,i))
        
```
